```python
features=["genres","keywords","title","tagline","overview","cast","director"]
```

```python
for feature in features :
  df[feature]=df[feature].fillna("")
```

```python
df_combined=df["genres"]+df["keywords"]+df["title"]+df["tagline"]+df["overview"]+df["cast"]+df["director"]
```

```python
df_combined
```

```python
# vectorization
verctorizer=TfidfVectorizer()
x=verctorizer.fit_transform(df_combined)
x.shape
```

```python
# calculate the cosine similarity
cos=cosine_similarity(x)
cos.shape
```

```python
movies=list(df.title)
movie_name=input("enter a movie name :")
matches=difflib.get_close_matches(movie_name,movies)
matches
```

```python
movie_name=matches[0]
movie_id=df[df["title"]==movie_name].index.values[0]
similarity_score=list(enumerate(cos[movie_id]))
similarity_score=sorted(similarity_score,key=lambda x :x[1],reverse=True)
similarity_score[:10]
```

```python
for i in range(10):
  movie_idx,_=similarity_score[i]
  similar_movie=df.loc[movie_idx,"title"]
  print(f"{i+1} . {similar_movie}")
```
